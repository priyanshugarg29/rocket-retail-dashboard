### Missing Values Analysis

The goal here is to:

- Calculate the total number of missing values in each dataset.

- Summarise how many missing values exist in each column.

- Validate whether these missing values are expected (e.g., transactionid in non-purchase events, parentid in root categories) or need attention.

This step will help determine whether additional data cleaning is needed before analysis.

```python
# Function to print total and column-wise missing values in a DataFrame

def print_missing_values_summary(dataframe, dataframe_name):
    total_missing = dataframe.isnull().sum().sum()
    missing_per_column = dataframe.isnull().sum()

    print(f"Missing value summary for {dataframe_name}:")
    print(f"Total records in {dataframe_name}: {dataframe.shape[0]}")
    print(f"Total missing values in {dataframe_name}: {total_missing}\n")

    if total_missing == 0:
        print("No missing values found in any column.\n")
    else:
        print("Missing values by column:")
        print(missing_per_column[missing_per_column > 0])
        print("-" * 30)

# Running for each dataset
print_missing_values_summary(df_events, 'df_events')
print_missing_values_summary(df_item_properties, 'df_item_properties')
print_missing_values_summary(df_category_tree, 'df_category_tree')
```

### Time Gap Analysis Between Events

In this step, we will:

- Calculate the time difference between consecutive events for each visitor.

- Convert these differences into seconds.

- Analyse the distribution of time gaps to help determine a sensible session timeout threshold (e.g., 30 minutes or 1 hour).

Visualise the results to understand how frequently users return within the same session.

```python
# Sorting events by visitor and timestamp to calculate time gaps correctly
df_events_sorted = df_events.sort_values(['visitorid', 'timestamp']).copy()

# Calculating time difference between consecutive events per visitor
df_events_sorted['time_diff'] = df_events_sorted.groupby('visitorid')['timestamp'].diff()

# Removing NaN time differences (which occur for the first event of each visitor)
df_events_sorted = df_events_sorted.dropna(subset=['time_diff'])

# Converting time differences from milliseconds to seconds
df_events_sorted['time_diff_sec'] = df_events_sorted['time_diff'] / 1000

# Taking the log scale for better visualisation (many small gaps, some very large)
df_events_sorted['log_time_gap'] = np.log1p(df_events_sorted['time_diff_sec'])

# Saving the dataframe
save_summary(df_events_sorted, 'df_events_sorted.csv')
```

This approach balances the natural breakpoints in user behaviour observed in the data and aligns with common practices in session-based modelling.

### Session Construction Function

```python
# Function to assign session IDs based on time gaps
def assign_sessions(dataframe, time_gap_threshold_seconds=1800):
    dataframe = dataframe.sort_values(['visitorid', 'timestamp']).copy()

    # Calculate time difference between consecutive events
    dataframe['time_diff_sec'] = dataframe.groupby('visitorid')['timestamp'].diff().fillna(0) / 1000

    # Start new session if time gap > threshold
    dataframe['new_session'] = dataframe['time_diff_sec'] > time_gap_threshold_seconds

    # Assign session IDs using cumulative sum
    dataframe['session_id'] = dataframe.groupby('visitorid')['new_session'].cumsum()

    # Create unique session ID combining visitor and session number
    dataframe['session_id'] = dataframe['visitorid'].astype(str) + "_" + dataframe['session_id'].astype(str)

    return dataframe

# Applying the session construction
df_sessions = assign_sessions(df_events, time_gap_threshold_seconds=1800)

# Checking the first few rows
print("Sample rows with session IDs assigned:")
print(df_sessions[['visitorid', 'timestamp', 'event', 'session_id']].head(10))

# Saving Sessions
save_summary(df_sessions, 'df_sessions.csv')
```

## Session Construction

```python
# Defining session timeout threshold (30 minutes in seconds)
session_timeout_sec = 1800

# Sorting events again for safety
df_events_sorted = df_events.sort_values(['visitorid', 'timestamp']).copy()

# Calculating time difference between events per visitor
df_events_sorted['time_diff'] = df_events_sorted.groupby('visitorid')['timestamp'].diff() / 1000  # converting to seconds

# Identifying where a new session should begin
df_events_sorted['new_session'] = (df_events_sorted['time_diff'] > session_timeout_sec) | (df_events_sorted['time_diff'].isnull())

# Cumulative session ID per visitor
df_events_sorted['session_id'] = df_events_sorted.groupby('visitorid')['new_session'].cumsum()

# Optional: create a global unique session ID
df_events_sorted['session_global_id'] = df_events_sorted['visitorid'].astype(str) + '_' + df_events_sorted['session_id'].astype(str)

# Verifying the output
print("Sample sessionised events:")
print(df_events_sorted[['visitorid', 'timestamp', 'event', 'session_id', 'session_global_id']].head(10))
```

To convert raw event logs into structured behaviour sessions, we applied a session timeout threshold of 30 minutes. Events were sorted by visitor and timestamp, and time gaps between consecutive actions were calculated. A new session was initiated whenever a user was inactive for more than 1,800 seconds or when the first event for a visitor occurred.

Each session was assigned a unique session_id per visitor, and a global session identifier (session_global_id) was created for easier analysis and embedding preparation. This session structure will form the foundation for downstream modelling, where the goal is to capture meaningful behavioural sequences rather than isolated events.

## Conversion Funnel Breakdown: From Views to Transactions

A critical component of understanding ecommerce behaviour is tracking how users progress through the funnel—from viewing a product, to adding it to cart, to completing a transaction.

This section quantifies and visualizes the flow of users across these three interaction types. By plotting drop-offs between stages, we highlight frictions in the customer journey and surface opportunities to enhance the conversion rate th
rough interventions like retargeting, UX improvements, or personalised offers.

```python
# Computing total number of each event type
event_counts = df_events['event'].value_counts().to_dict()

# Ensuring all stages are represented
views = event_counts.get('view', 0)
add_to_carts = event_counts.get('addtocart', 0)
transactions = event_counts.get('transaction', 0)

# Preparing funnel data
funnel_values = [views, add_to_carts, transactions]
funnel_labels = ['View', 'Add to Cart', 'Transaction']

# Plotting: Horizontal funnel chart
plt.figure(figsize=(10, 5))
bars = plt.barh(funnel_labels, funnel_values, color=['skyblue', 'orange', 'seagreen'])

# Annotating values on bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 1000, bar.get_y() + bar.get_height() / 2,
             f'{width:,.0f}', va='center', fontsize=10)

plt.title('Conversion Funnel: View ➝ Add to Cart ➝ Transaction', fontsize=14)
plt.xlabel('Number of Events')
plt.tight_layout()

# Saving for Streamlit dashboard
plt.savefig('conversion_funnel.png')
plt.show()
```

- Out of **2.66 million product views**, only **69,332** proceed to add-to-cart, and just **22,457** lead to an actual transaction.
- This reveals a **conversion rate of ~2.6%** from view to add-to-cart, and **~32.4%** from add-to-cart to transaction.
- The **largest drop-off** occurs at the exploration stage, between viewing and carting, indicating potential barriers such as insufficient product information, pricing hesitancy, or a lack of purchase urgency.
- Despite the steep drop-off, users who do engage with the cart show a strong **conversion signal**, suggesting this cohort may be particularly responsive to **nudges, retargeting, or incentives**.
- These insights support the use of **behaviour-triggered campaigns**, such as abandoned cart reminders, exit-intent modals, and targeted promotions to strengthen funnel efficiency.

## Basket Size Analysis: Quantity of Items per Purchase

Basket size represents the number of items a customer purchases during a single transaction. Understanding this metric is essential to gauge average order behavior, identify bundling opportunities, and develop targeted incentives for increasing cart value.

In this section, we estimate basket sizes by grouping transactions by `visitorid` and `timestamp` at a daily level. Although the Rocket Retail dataset lacks explicit transaction session IDs, this approximation offers useful insights into how many items users typically buy at once.


```python
# Filtering only transaction events
transaction_df = df_events[df_events['event'] == 'transaction'].copy()

# Ensuring timestamp is datetime (necessary after filtering)
transaction_df['timestamp'] = pd.to_datetime(transaction_df['timestamp'])

# Approximating baskets using visitorid and date
transaction_df['date'] = transaction_df['timestamp'].dt.date
basket_counts = transaction_df.groupby(['visitorid', 'date']).size()

# Summary statistics
basket_summary = basket_counts.describe()
print("Basket Size Summary:\n", basket_summary)

# Plotting distribution of basket sizes
plt.figure(figsize=(10, 5))
plt.hist(basket_counts, bins=30, color='mediumpurple', edgecolor='black')
plt.title('Distribution of Basket Sizes (Items per Purchase)', fontsize=14)
plt.xlabel('Number of Items in Basket')
plt.ylabel('Number of Purchases')
plt.tight_layout()

# Saving for Streamlit dashboard
plt.savefig('basket_size_distribution.png')
plt.show()
```

- The **median basket size is just 1 item**, indicating that the majority of customers complete purchases with only a single product.
- Despite this, the **maximum observed basket size exceeds 500 items**, highlighting a few large-volume purchases likely associated with bulk-buying behavior or system-level anomalies.
- With a **mean basket size of ~1.9**, the distribution is **highly right-skewed**, with most purchases being minimal and a long tail of larger transactions.
- These patterns suggest opportunities for:
  - **Cross-sell and bundling strategies** to encourage multi-item carts
  - **Free shipping thresholds** to incentivize larger baskets
  - **Flagging outlier behavior** for validation or high-value customer tracking
- Understanding basket sizes helps tailor conversion-focused strategies for both low and high-volume buyers.

## Event Lag Analysis: Time Between View ➝ Cart ➝ Purchase

Understanding how much time users take to move from viewing a product to adding it to cart — and eventually purchasing — reveals patterns of intent, hesitation, and opportunity for real-time interventions.

This section calculates the time lags between `view`, `addtocart`, and `transaction` events for the same user and item. By analyzing the delay distribution, we can identify whether users act instantly or deliberate over time, and tailor campaign timing accordingly.


```python
# Filtering for relevant events only
df_lag = df_events[df_events['event'].isin(['view', 'addtocart', 'transaction'])].copy()
df_lag['timestamp'] = pd.to_datetime(df_lag['timestamp'])

# Sorting for lag calculation
df_lag = df_lag.sort_values(['visitorid', 'itemid', 'timestamp'])

# Assigning numerical code to each event type
df_lag['event_code'] = df_lag['event'].map({'view': 0, 'addtocart': 1, 'transaction': 2})

# For each visitor-item pair, computing lags between view → cart → transaction
lags = []

for (visitor, item), group in df_lag.groupby(['visitorid', 'itemid']):
    events = group.sort_values('timestamp')
    timestamps = events['timestamp'].tolist()
    codes = events['event_code'].tolist()

    view_time = cart_time = trans_time = None
    for t, code in zip(timestamps, codes):
        if code == 0:
            view_time = t
        elif code == 1 and view_time:
            cart_time = t
        elif code == 2 and cart_time:
            trans_time = t

        # Once all 3 exist, computing lags in seconds
        if view_time and cart_time and trans_time:
            lags.append({
                'visitorid': visitor,
                'itemid': item,
                'view_to_cart_secs': (cart_time - view_time).total_seconds(),
                'cart_to_trans_secs': (trans_time - cart_time).total_seconds()
            })
            break

# Creating DataFrame from lag results
lag_df = pd.DataFrame(lags)

# Plotting histograms
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(lag_df['view_to_cart_secs'], bins=50, color='goldenrod', edgecolor='black')
plt.title('Lag: View ➝ Add to Cart (Seconds)')
plt.xlabel('Seconds')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(lag_df['cart_to_trans_secs'], bins=50, color='teal', edgecolor='black')
plt.title('Lag: Add to Cart ➝ Transaction (Seconds)')
plt.xlabel('Seconds')
plt.ylabel('Frequency')

plt.tight_layout()
plt.savefig('event_lag_analysis_seconds.png')
plt.show()
```
